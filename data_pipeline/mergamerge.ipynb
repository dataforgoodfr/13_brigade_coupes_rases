## Mega Merge


```python
# TODO: how to automatically choose the right number of partitions?
sufosat_dask = dask_geopandas.from_geopandas(sufosat, npartitions=12)
sufosat_dask
```

```python
def overlay(
    left_gdf: dask_geopandas.GeoDataFrame, right_gdf: gpd.GeoDataFrame
) -> dask_geopandas.GeoDataFrame:
    """
    dask-geopandas doesn't come with the `overlay` function so we make our own.

    """
    # Perform spatial join between left and right GeoDataFrames
    right_gdf = right_gdf.assign(right_geometry=right_gdf.geometry)

    # Spatial join the GeoDataFrames (using left and right geometries)
    # dask-geopandas only supports "inner" joins
    joined_gdf = left_gdf.sjoin(right_gdf, how="inner", predicate="intersects")

    # Calculate intersection of geometries
    joined_gdf["geometry"] = joined_gdf.geometry.intersection(joined_gdf.right_geometry)

    # Drop temporary right_geometry column
    joined_gdf = joined_gdf.set_geometry(col="geometry").drop(columns="right_geometry")

    return joined_gdf
```

### Cities


```python
# TODO: Add departments
cities = (
    overlay(sufosat_dask, cadastre_cities)
    .groupby("clear_cut_group")["code_insee"]
    .agg(list)
    .compute()
)
cities
```

```python
sufosat.loc[cities.index, "cities"] = cities
sufosat["cities"].isna().sum()
```

```python
# Some clear-cut clusters don't fall within any city polygon
# Clear-cuts without any city intersection are joined with their nearest cities
# (can be multiple cities if they are equidistant to the clear-cut)
clear_cuts_with_nearest_cities = (
    sufosat[sufosat["cities"].isna()]
    .sjoin_nearest(cadastre_cities)
    .groupby("clear_cut_group")["code_insee"]
    .agg(list)
)
clear_cuts_with_nearest_cities
```

```python
sufosat.loc[clear_cuts_with_nearest_cities.index, "cities"] = clear_cuts_with_nearest_cities
```

### Natura 2000


```python
# 1 hectare = 10,000 m²
natura2000_area_ha: "pd.Series[np.floating]" = (
    overlay(sufosat_dask, natura2000).area.compute() / 10000
)
natura2000_area_ha
```

```python
sufosat.loc[natura2000_area_ha.index, "natura2000_area_ha"] = natura2000_area_ha
sufosat
```

### Slope


```python
# Get the geometry of the intersection between SUFOSAT and the slopes
slope_intersection = overlay(sufosat_dask, dask_geopandas.from_geopandas(slope, npartitions=12))

# Since we want the area of the single large continuous slope polygon
# we explode multi-polygons into simple polygons
slope_intersection = slope_intersection.explode()

# Compute the simple polygons intersection area
# 1 hectare = 10,000 m²
slope_area_ha: "pd.Series[np.floating]" = slope_intersection.area.compute() / 10000

# A clear-cut cluster can intersect with several slope polygons. In our case,
# we want to keep the single largest slope intersection area for each cluster
slope_area_ha = slope_area_ha.sort_values(ascending=True)
slope_area_ha = slope_area_ha[~slope_area_ha.index.duplicated(keep="last")]

slope_area_ha
```

```python
sufosat.loc[slope_area_ha.index, "slope_area_ha"] = slope_area_ha
sufosat
```
