```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
df = gpd.read_file("zps.shp")
df.head()
```

```python
print(df.crs)
```

```python
df["Aire_m2"] = df.geometry.area
df["Aire_km2"] = df["Aire_m2"] / 1e6
```

```python
df.head()
```

```python
df2 = gpd.read_file("sic.shp")
df2.head()
```

```python
print(df2.crs)
```

```python
df2["Aire_m2"] = df2.geometry.area
df2["Aire_km2"] = df2["Aire_m2"] / 1e6
df2.head()
```

```python
nb_communs = len(set(df["SITENAME"]) & set(df2["SITENAME"]))
print(f"Nombre d'éléments communs : {nb_communs}")
```

```python
df.info()
```

```python
df2.info()
```

```python
df["Couches"] = "ZPS"
df2["Couches"] = "ZSC"
```

```python
df.head()
```

```python
df_merged = pd.concat([df, df2], ignore_index=True)
```

```python
df_merged.head()
```

```python
df_merged.info()
```

```python
df_light = df_merged[["SITECODE", "SITENAME", "Aire_km2", "Couches"]]
df_light.head()
```
